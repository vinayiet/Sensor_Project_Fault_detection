```python
def get_col_with_std_dev(df):
    cols_to_drop = []
    
    # Loop through columns, excluding non-numeric columns
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']:  # Check if the column is numeric
            if df[col].std() == 0:  # Assuming you are looking for columns with zero std dev
                cols_to_drop.append(col)
        else:
            print(f"Skipping non-numeric column: {col}")
    
    return cols_to_drop
```

```python
def get_redundant_col(df: pd.DataFrame, missing_threshold = .7):
    cols_missing_ratio = df.isna().sum().div(df.shape[0])
    cols_to_drop = cols_missing_ratio[cols_missing_ratio > missing_threshold].index
    return cols_to_drop
```
